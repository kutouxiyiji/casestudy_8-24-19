```

df_Time= pd.DataFrame(df_RMA.groupby(by=['Time','SKU'],as_index=False)['Dollar Sales', 'Unit Sales'].sum())
df_Time.head()
```

```
df_Time['Time'] = pd.to_datetime(df_Time.Time)
```

```
df_Time.sort_values(by = 'Time', ascending=True)
```

```
def df_Time_plot(SKU,x,y,df = df_Time ):
  fig = df_Time[df_Time['SKU']==SKU].plot(x=x,y=y,figsize  = (15,10),fontsize=20)
  fig.set_ylabel(y, fontsize=20)
  fig.set_xlabel(x, fontsize=20)
  fig.set_title(SKU + ' over ' + x,fontsize = 25)
  fig.legend(loc='upper left', fontsize = 20)
  plt.tight_layout()
  #plt.show()
```

```
df_Time_plot('E-Cigs Total','Time','Unit Sales')
df_Time_plot('Cigarettes Total','Time','Unit Sales')
df_Time_plot('JUUL Devices','Time','Unit Sales')
df_Time_plot('JUUL Refill Kits','Time','Unit Sales')
#df_Time[df_Time['SKU']=='Cigarettes Total'].plot(x='Time',y='Unit Sales')
#need add accessories
```

```
df_Time_JUULrefill = df_Time[df_Time['SKU'] == 'JUUL Refill Kits']
df_Time_JUULrefill = pd.DataFrame(df_Time_JUULrefill)
df_Time_JUULrefill.head()
```

```
df_Time_JUULrefill = df_Time_JUULrefill.sort_values(by = 'Time', ascending=True)
df_Time_JUULrefill = df_Time_JUULrefill.reset_index(drop=True)
df_Time_JUULrefill['Unit Sales PCTchange'] = df_Time_JUULrefill['Unit Sales'].pct_change()
#df_Time_JUULrefill['Time'] = pd.to_datetime(df_Time_JUULrefill.Time)

df_Time_JUULrefill.head()
```

```
def general_line_plot(x,y,df,title = ''):
  fig = df.plot(x=x,y=y,figsize  = (15,10),fontsize=20)
  fig.set_ylabel(y, fontsize=20)
  fig.set_xlabel(x, fontsize=20)
  fig.set_title(title,fontsize = 25)
  fig.legend(loc='upper left', fontsize = 20)
  plt.tight_layout()
```

```
general_line_plot('Time','Unit Sales PCTchange',df_Time_JUULrefill,'Juul refill kits Unit Sales pecent change over time')
```

```
def df_add_pct_change(category, subcategory, Sales , df):
  df_res = df[df[category] == subcategory]
  df_res = pd.DataFrame(df_res)
  df_res = df_res.sort_values(by = 'Time', ascending = True)
  df_res = df_res.reset_index(drop=True)
  df_res[Sales + ' PCTchange'] = df_res[Sales].pct_change() * 100
  return df_res
```

```
df_Time_ECigs = df_add_pct_change('SKU','E-Cigs Total', 'Unit Sales', df_Time)
general_line_plot('Time','Unit Sales PCTchange',df_Time_ECigs,title = 'E-cigs total PCT change of Unit sales over time')
```

```
# double check the df_add_pct_change function is right
df_Time_JUULrefill2 = df_add_pct_change('SKU','JUUL Refill Kits', 'Unit Sales', df_Time)
general_line_plot('Time','Unit Sales PCTchange',df_Time_JUULrefill2 ,title = 'Juul refill kits PCT change of Unit sales over time')
```

```
def three_df_overlay_line_plot(x,y1,df1,y2,df2,y3 = None,df3=None, title = ''):
  ax = df1.plot(x=x,y=y1,figsize  = (15,10),fontsize=20)
  df2.plot(ax=ax,x=x,y=y2,figsize  = (15,10),fontsize=20)
  if y3:
    df3.plot(ax=ax,x=x,y=y3,figsize  = (15,10),fontsize=20)
  ax.set_ylabel(y1, fontsize=20)
  ax.set_xlabel(x, fontsize=20)
  ax.set_title(title,fontsize = 25)
  if y3:
    ax.legend([df1.SKU[0] + ' ' + y1, df2.SKU[0] +' ' +y2, df3.SKU[0] +' ' +y3],loc='upper left', fontsize = 20)  
  else:
    ax.legend([df1.SKU[0] + ' ' + y1, df2.SKU[0] +' ' +y2],loc='upper left', fontsize = 20)
  plt.tight_layout()
   
  
```

```
three_df_overlay_line_plot('Time','Unit Sales PCTchange', df_Time_JUULrefill,'Unit Sales PCTchange', df_Time_ECigs,title = 'JUUL refill kits vs. E-cigs total PCT change over time')
```

```
# here we create all the different DataFrames which include Unit PCT change and Dollors PCT change over time for different SKU
#first cigreates total
df_Time_Cigarettes_Total_PCT_change = df_add_pct_change('SKU','Cigarettes Total', 'Unit Sales', df_Time)
df_Time_Cigarettes_Total_PCT_change = df_add_pct_change('SKU','Cigarettes Total', 'Dollar Sales', df_Time_Cigarettes_Total_PCT_change)

#E-cigs total
df_Time_ECigs_Total_PCT_change = df_add_pct_change('SKU','E-Cigs Total', 'Unit Sales', df_Time)
df_Time_ECigs_Total_PCT_change = df_add_pct_change('SKU','E-Cigs Total', 'Dollar Sales', df_Time_ECigs_Total_PCT_change)
#JUUL Refill Kits
df_Time_JUUL_Refill_Kits_PCT_change = df_add_pct_change('SKU','JUUL Refill Kits', 'Unit Sales', df_Time)
df_Time_JUUL_Refill_Kits_PCT_change = df_add_pct_change('SKU','JUUL Refill Kits', 'Dollar Sales', df_Time_JUUL_Refill_Kits_PCT_change)
#JUUL Devices
df_Time_JUUL_Devices_PCT_change = df_add_pct_change('SKU','JUUL Devices', 'Unit Sales', df_Time)
df_Time_JUUL_Devices_PCT_change = df_add_pct_change('SKU','JUUL Devices', 'Dollar Sales', df_Time_JUUL_Devices_PCT_change)

# check the df head
df_Time_Cigarettes_Total_PCT_change.head() ,df_Time_ECigs_Total_PCT_change.head() ,df_Time_JUUL_Refill_Kits_PCT_change.head() ,df_Time_JUUL_Devices_PCT_change.head()
```

```
# first we plot the cigs total vs. E-cigs total PCT change over time, both Unit Sales/ Dollar Sales
three_df_overlay_line_plot('Time','Unit Sales PCTchange', df_Time_Cigarettes_Total_PCT_change,'Unit Sales PCTchange', df_Time_ECigs_Total_PCT_change ,title = 'Cigs Total vs. E-Cigs Total PCT change over time (Unit Sales)')
three_df_overlay_line_plot('Time','Dollar Sales PCTchange', df_Time_Cigarettes_Total_PCT_change,'Dollar Sales PCTchange', df_Time_ECigs_Total_PCT_change ,title = 'Cigs Total vs. E-Cigs Total PCT change over time (Dollar Sales)')
```
