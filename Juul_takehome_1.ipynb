```
df = pd.read_csv('/content/drive/My Drive/Python/dataSet/Juul/Case_Study_Retail_Data_clean_1.csv')
```

```
df['SKU'].describe(), df['SKU'].unique()
```

```
df_SKU = df.groupby(['SKU']).sum()
df_SKU
```

```
df_SKU.loc['E-Cigs Total']
```

```
fig1, fig2 = df_SKU.plot.bar(subplots=True, y = ['Dollar Sales','Unit Sales'],log=True, figsize  = (20,15),fontsize=20)
fig1.title.set_size(20)
fig2.title.set_size(20)
fig1.set_ylabel('Dollar Sales log scale per SKU', fontsize = 20)
fig2.set_ylabel('Unit Sales log scale per SKU', fontsize=20)
fig2.set_xlabel('')
#plt.legend(loc='left center', bbox_to_anchor=(-0.1, 1.), fontsize=8)
#axs[0].set_title('subplot 1')
#plt.tight_layout()

#fig.legend(fontsize)
```

```
fig1, fig2 = df_SKU.loc[['Cigarettes Total','E-Cigs Total']].plot.bar(subplots=True, y = ['Dollar Sales','Unit Sales'], figsize  = (20,15),fontsize=20)
fig1.title.set_size(20)
fig2.title.set_size(20)
fig1.set_ylabel('Dollar Sales', fontsize = 20)
fig2.set_ylabel('Unit Sales', fontsize=20)
fig2.set_xlabel('')
plt.tight_layout()
```

```
fig1,fig2 = df_SKU.loc[['Cigarettes Total','E-Cigs Total']].plot.pie(subplots=True,figsize  = (20,15),fontsize=20)
fig1.set_title('Dollar Sales',fontsize = 20)
fig1.set_ylabel('')
fig2.set_title('Unit Sales',fontsize = 20)
fig2.set_ylabel('')
plt.tight_layout()
```
